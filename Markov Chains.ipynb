```python
import numpy as np
import random as rm

states = ["v1","v2","v6","v7"]
transitions = [["v1","v2","v6","v7"],["v1","v2","v6","v7"],["v1","v2","v6","v7"],["v1","v2","v6","v7"]]
T = [[0.5,0.3,0.2,0],[0.1,0.4,0.1,0.4],[0.2,0.4,0.2,0.2],[0.1,0.2,0.4,0.3]]



def forecast(times, cur_state="v1"):
    state_list=[cur_state]
    n = 1
    prob = 1.0
    time_counter = [0,0,0,0]
    if cur_state == "v1":
        time_counter[0]+=1
    elif cur_state == "v2":
        time_counter[1]+=1
    elif cur_state == "v6":
        time_counter[2]+=1
    elif cur_state == "v7":
        time_counter[3]+=1
    while n != times:
        if cur_state == "v1":
            change = np.random.choice(transitions[0],replace=True, p=T[0])
            if change == "v1":
                prob*=T[0][0]
                state_list.append("v1")
            elif change == "v2":
                prob*=T[0][1]
                state_list.append("v2")
            elif change == "v6":
                prob*=T[0][2]
                state_list.append("v6")
            elif change == "v7":
                prob*=T[0][3]
                state_list.append("v7")
            cur_state = change
            time_counter[0]+=1
        elif cur_state == "v2":
            change = np.random.choice(transitions[1],replace=True, p=T[1])
            if change == "v1":
                prob*=T[1][0]
                state_list.append("v1")
            elif change == "v2":
                prob*=T[1][1]
                state_list.append("v2")
            elif change == "v6":
                prob*=T[1][2]
                state_list.append("v6")
            elif change == "v7":
                prob*=T[1][3]
                state_list.append("v7")
            cur_state = change
            time_counter[1]+=1
        elif cur_state == "v6":
            change = np.random.choice(transitions[2],replace=True, p=T[2])
            if change == "v1":
                prob*=T[2][0]
                state_list.append("v1")
            elif change == "v2":
                prob*=T[2][1]
                state_list.append("v2")
            elif change == "v6":
                prob*=T[2][2]
                state_list.append("v6")
            elif change == "v7":
                prob*=T[2][3]
                state_list.append("v7")
            cur_state = change
            time_counter[2]+=1
        elif cur_state == "v7":
            change = np.random.choice(transitions[3],replace=True, p=T[3])
            if change == "v1":
                prob*=T[3][0]
                state_list.append("v1")
            elif change == "v2":
                prob*=T[3][1]
                state_list.append("v2")
            elif change == "v6":
                prob*=T[3][2]
                state_list.append("v6")
            elif change == "v7":
                prob*=T[3][3]
                state_list.append("v7")
            cur_state = change
            time_counter[3]+=1
        n+=1
    time_percent = [time / n for time in time_counter]
    return [state_list, time_counter, time_percent]
    
```

## Цепь длины 10

```python
res1_10 = forecast(10, "v1")
res2_10 = forecast(10, "v2")
res6_10 = forecast(10, "v6")
res7_10 = forecast(10, "v7")
```
